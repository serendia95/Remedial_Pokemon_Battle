# Pre-processing

```python
pokemon1 = pokemon.drop(["Type 1", "Type 2", "Generation", "Legendary", "Name"], axis = "columns")
pokemon1.head()
```

```python
pokemon1.isna().sum()
```

```python
pokemon2 = pd.merge(combating, pokemon1, left_on = "First_pokemon", right_on = "#", how = "inner")
```

```python
pokemon2.head()
```

```python
pokemon2 = pokemon2.rename({"First_pokemon" : "Pokemon_A", "Second_pokemon" : "Pokemon_B", "#" : "ID", "HP" : "HP_A",
                            "Attack" : "Attack_A", "Defense" : "Defense_A", "Sp. Atk" : "Sp. Atk_A", "Sp. Def" : "Sp. Def_A",
                            "Speed" : "Speed_A"}, axis = "columns")
```

```python
pokemon2.head()
```

```python
pokemon2 = pd.merge(pokemon2, pokemon1, left_on = "Pokemon_B", right_on = "#", how = "inner")
pokemon2 = pokemon2.drop(['#', 'ID'], axis = "columns")
```

```python
pokemon2 = pokemon2.rename({"HP" : "HP_B", "Attack" : "Attack_B", "Defense" : "Defense_B", "Sp. Atk" : "Sp. Atk_B", 
                            "Sp. Def" : "Sp. Def_B", "Speed" : "Speed_B"}, axis = "columns")
```

```python
# Checking
pokemon3 = pokemon2[pokemon2["Pokemon_A"] == 31]
pokemon3 = pokemon3[pokemon3["Pokemon_B"] == 163]
pokemon3
```

```python
win = []

for i in range(len(pokemon2)):
    if pokemon2["Winner"].iloc[i] == pokemon2["Pokemon_A"].iloc[i]:
        h = 1
        win.append(h)
    else:
        h = 0
        win.append(h)
        
pokemon2["Winner"] = win
```

```python
pokemon2.head()
```

# Modelling

```python
from sklearn.ensemble import RandomForestClassifier
x = pokemon2.drop(["Pokemon_A", "Pokemon_B", "Winner"], axis = "columns")
y = pokemon2["Winner"]

from sklearn.model_selection import train_test_split
xtr, xts, ytr, yts = train_test_split(x, y, test_size = 0.18)

model_rfc = RandomForestClassifier(n_estimators = 100)
model_rfc.fit(xtr, ytr)
score_rfc = model_rfc.score(xts, yts)
```

# Scoring

```python
y_pred = model_rfc.predict(xts)
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import classification_report
from sklearn.metrics import jaccard_score, jaccard_similarity_score
from sklearn.metrics import roc_curve, roc_auc_score, auc

print(f"Accuracy Score: {score_rfc}")
print(f"Balance Accuracy Score: {round(balanced_accuracy_score(yts, y_pred), 2)}")
print(f"Precision Score: {precision_score(yts, y_pred)}")
print(f"Recall Score: {recall_score(yts, y_pred)}")
print(f"F1 Score: {f1_score(yts, y_pred)}")
print(f"ROC AUC Score: {round(roc_auc_score(yts,y_pred),2)}")
```

# Prediction and Finalization

```python
def battle(a, b):
    
    a = a.split(" ")
    a1 = []
    for i in a:
        a1.append(i.capitalize())
    a = " ".join(a1)
        
    b = b.split(" ")
    b1 = []
    for i in b:
        b1.append(i.capitalize())
    b = " ".join(b1)
    
    try:
    
        ID_A = pokemon["#"][pokemon["Name"] == a].values[0]
        HP_A = pokemon["HP"][pokemon["Name"] == a].values[0]
        Attack_A = pokemon["Attack"][pokemon["Name"] == a].values[0]
        Defense_A = pokemon["Defense"][pokemon["Name"] == a].values[0]
        Sp_Atk_A = pokemon["Sp. Atk"][pokemon["Name"] == a].values[0]
        Sp_Def_A = pokemon["Sp. Def"][pokemon["Name"] == a].values[0]
        Speed_A = pokemon["Speed"][pokemon["Name"] == a].values[0]

        ID_B = pokemon["#"][pokemon["Name"] == b].values[0]
        HP_B = pokemon["HP"][pokemon["Name"] == b].values[0]
        Attack_B = pokemon["Attack"][pokemon["Name"] == b].values[0]
        Defense_B = pokemon["Defense"][pokemon["Name"] == b].values[0]
        Sp_Atk_B = pokemon["Sp. Atk"][pokemon["Name"] == b].values[0]
        Sp_Def_B = pokemon["Sp. Def"][pokemon["Name"] == b].values[0]
        Speed_B = pokemon["Speed"][pokemon["Name"] == b].values[0]

        prediction = model_rfc.predict([[HP_A, Attack_A, Defense_A, Sp_Atk_A, Sp_Def_A, Speed_A, 
                                         HP_B, Attack_B, Defense_B, Sp_Atk_B, Sp_Def_B, Speed_B]])
        probability = model_rfc.predict_proba([[HP_A, Attack_A, Defense_A, Sp_Atk_A, Sp_Def_A, Speed_A, 
                                                HP_B, Attack_B, Defense_B, Sp_Atk_B, Sp_Def_B, Speed_B]])
        max_probability = probability[0].max()*100

        if prediction == 0:
            winner = b
        elif prediction == 1:
            winner = a

        return print(f"Winner: {int(max_probability)}% {winner}")
    
    except:
        
        return print("Maaf, data tidak ditemukan.")
```

```python
battle("PIKACHU", "MEWTWO")
battle("PIKACHU", "CHARIZARD")
battle("CHARMANDER", "DITTO")
battle("MEGA CHARIZARD X", "CHARIZARD")
battle("AGUMON", "PATAMON")
```
